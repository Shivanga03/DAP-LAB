```python
import pandas as pd
import numpy as np
```

```python
weather_data= pd.read_csv("weather.csv")
```

```python
weather_data.head()
```

```python
#  Find all the unique 'Wind Speed' values in the data

unique_wind_speeds = weather_data['Wind Speed_km/h'].unique()
print(unique_wind_speeds)
```

```python
# Find the number of times when the 'Weather is exactly Clear'.

clear_weather_count = (weather_data['Weather'] == 'Clear').sum()

print("Number of times when the Weather is exactly Clear:", clear_weather_count)
```

```python
# Find the number of times when the 'Wind Speed was exactly 4 km/h'.

wind_speed_4_count = (weather_data['Wind Speed_km/h'] == 4).sum()
print("Number of times when the Wind Speed was exactly 4 km/h:", wind_speed_4_count)
```

```python
# Find out all the Null Values in the data.

null_values = weather_data.isnull().sum()
print("Null values in the data:\n", null_values)
```

```python
# Rename the column name 'Weather' of the dataframe to 'Weather Condition'.

weather_data.rename(columns={'Weather': 'Weather Condition'}, inplace=True)
print("Column 'Weather' renamed to 'Weather Condition'.")
```

```python
# What is the mean 'Visibility' ?

mean_visibility = weather_data['Visibility_km'].mean()
print("Mean Visibility:", mean_visibility)
```

```python
# What is the Standard Deviation of 'Pressure'  in this data?

pressure_std_deviation = weather_data['Press_kPa'].std()
print("Standard Deviation of Pressure:", pressure_std_deviation)
```

```python
# What is the Variance of 'Relative Humidity' in this data ?


humidity_variance = weather_data['Rel Hum_%'].var()
print("Variance of Relative Humidity:", humidity_variance)
```

```python
# Find all instances when 'Snow' was recorded.

snow_records = weather_data[weather_data['Weather Condition'].str.contains('Snow')]
print("Instances when 'Snow' was recorded:\n", snow_records)
```

```python
# Find all instances when 'Wind Speed is above 24' and 'Visibility is 25'.

high_wind_visibility = weather_data[(weather_data['Wind Speed_km/h'] > 24) & (weather_data['Visibility_km'] == 25)]
print("Instances when Wind Speed is above 24 and Visibility is 25:\n", high_wind_visibility)
```

```python
# What is the Mean value of each column against each 'Weather Condition ?

mean_by_weather_condition = weather_data.groupby('Weather Condition').mean()
print("Mean value of each column against each Weather Condition:\n", mean_by_weather_condition)
```

```python
# What is the Minimum & Maximum value of each column against each 'Weather Condition ?

min_max_by_weather_condition = weather_data.groupby('Weather Condition').agg(['min', 'max'])
print("Minimum & Maximum value of each column against each Weather Condition:\n", min_max_by_weather_condition)
```

```python
# Show all the Records where Weather Condition is Fog.

fog_records = weather_data[weather_data['Weather Condition'] == 'Fog']
print("Records where Weather Condition is Fog:\n", fog_records)
```

```python
# Find all instances when 'Weather is Clear' or 'Visibility is above 40'.

clear_or_high_visibility = weather_data[(weather_data['Weather Condition'] == 'Clear') | (weather_data['Visibility_km'] > 40)]
print("Instances when Weather is Clear or Visibility is above 40:\n", clear_or_high_visibility)
```

```python
# Find all instances when :
# A. 'Weather is Clear' and 'Relative Humidity is greater than 50'
# B. 'Visibility is above 40'

condition_15 = ((weather_data['Weather Condition'] == 'Clear') & (weather_data['Rel Hum_%'] > 50)) | (weather_data['Visibility_km'] > 40)
instances_15 = weather_data[condition_15]
print("Instances when Weather is Clear and Relative Humidity > 50 or Visibility > 40:\n", instances_15)
```
